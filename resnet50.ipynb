```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
from datetime import datetime
from PIL import Image
import os
from IPython.display import display, Image as _Imgdis

IMAGE_SIZE = [224, 224]

# Define dataset paths
train_path = "C:\\Users\\Dhanush\\Desktop\\training"
test_path = "C:\\Users\\Dhanush\\Desktop\\testing"

# Preprocess images 
def convert_image(image_path):
    img = Image.open(image_path)
    if img.mode in ('RGBA', 'LA') or (img.mode == 'P' and 'transparency' in img.info):
        img = img.convert("RGBA")
    if image_path.endswith('.webp'):
        img = img.convert("RGB")
        new_path = image_path.rsplit('.', 1)[0] + '.png'
        img.save(new_path)
        return new_path
    return image_path

# Display first 10 sample images
folder_name = 'oral_cancer'
folder = os.path.join(train_path, folder_name)
if os.path.exists(folder):
    onlybenignfiles = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]
    print("Working with {0} images".format(len(onlybenignfiles)))
    for i in range(min(10, len(onlybenignfiles))):
        img_path = convert_image(os.path.join(folder, onlybenignfiles[i]))
        display(_Imgdis(filename=img_path, width=240, height=240))
else:
    print("The specified directory does not exist:", folder)

# Load and preprocess ResNet50 model
resnet = ResNet50(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
for layer in resnet.layers:
    layer.trainable = False


x = Flatten()(resnet.output)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
prediction = Dense(1, activation='sigmoid')(x)  

#  model
model = Model(inputs=resnet.input, outputs=prediction)
model.summary()

# Compile the model 
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# Data augmentation
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.8, 1.2],
    channel_shift_range=0.2
)

test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)


train_set = train_datagen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'  # Binary classification
)

test_set = test_datagen.flow_from_directory(
    test_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'  # Binary classification
)

# Calculate steps per epoch and validation steps 
steps_per_epoch = train_set.samples // train_set.batch_size
validation_steps = test_set.samples // test_set.batch_size

# Define callbacks
checkpoint = ModelCheckpoint(
    filepath='my_model.keras',
    save_best_only=True,
    verbose=1
)
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    verbose=1,
    restore_best_weights=True
)

# Train the model 
start = datetime.now()
model_history = model.fit(
    train_set,
    validation_data=test_set,
    epochs=100,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    callbacks=[checkpoint, early_stopping],
    verbose=2
)
duration = datetime.now() - start
print("Training completed in time: ", duration)

# Plot training history
plt.plot(model_history.history['accuracy'], label='train accuracy')
plt.plot(model_history.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(model_history.history['loss'], label='train loss')
plt.plot(model_history.history['val_loss'], label='validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, f1_score

test_set.reset()

steps = int(np.ceil(test_set.samples / test_set.batch_size))  # Convert to int

y_pred = model.predict(test_set, steps=steps, verbose=1)
y_pred = np.round(y_pred).astype(int)  # Convert probabilities to binary (0 or 1)

y_true = test_set.classes

assert len(y_true) == len(y_pred), f"Mismatch: y_true length = {len(y_true)}, y_pred length = {len(y_pred)}"

class_labels = list(test_set.class_indices.keys())

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)
disp.plot(cmap=plt.cm.Blues, colorbar=True)
plt.title('Confusion Matrix')
plt.show()

# Print Confusion Matrix Details
print("\nConfusion Matrix Details:")
print(f"True Negative (TN): {cm[0, 0]} - Model correctly predicted 'benign'")
print(f"False Positive (FP): {cm[0, 1]} - Model wrongly predicted 'malignent' when it was 'benign'")
print(f"False Negative (FN): {cm[1, 0]} - Model wrongly predicted 'benign' when it was 'malignent'")
print(f"True Positive (TP): {cm[1, 1]} - Model correctly predicted 'malignent'")

# F1 Score
f1 = f1_score(y_true, y_pred)
print(f"\nF1 Score: {f1:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_labels))
```
